```python
le.fit_transform(char_data[['Genre']])
```

```python
char_data['Genre_L_enc'] = le.fit_transform(char_data[['Genre']])
char_data
```

```python
le.fit_transform(char_data[['Publisher']])
```

```python
char_data['Publisher_L_enc'] = le.fit_transform(char_data[['Publisher']])
char_data
```
